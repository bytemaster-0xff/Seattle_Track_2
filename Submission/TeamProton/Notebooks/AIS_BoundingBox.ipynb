```python
import csv
import os
import numpy as np
import pandas as pd

zone = 17
year = 2017
month = 12
day = 1
hour = 12
minute = 15

inputPath = './Split/AIS_'+str(year)+'_'+str(month)+'_'+str(day)+'_'+str(hour)+'_'+str(minute)+'_Zone'+str(zone)+'.csv'

tb = pd.read_csv(inputPath,
            index_col=None,
            header=0,
            parse_dates=[1],
            infer_datetime_format=True)
tb.columns = ['MMSI', 'BaseDateTime', 'LAT', 'LON', 'SOG', 'COG', 'Heading','VesselName', 'IMO', 'CallSign', 'VesselType', 'Status', 'Length','Width', 'Draft', 'Cargo']
               
columns = ['LAT','LON']

rowCount = tb.shape[0]
print(rowCount)    
blocks = dict();

for hour in range(23):
    for minuteIdx in range(4):
        minute = minuteIdx * 15
        print('Time', hour, minute)
        
        for i in range(rowCount):
            try:
                lat = int(tb["LAT"][i] + 90)
                lon = int(tb["LON"][i] + 90)
                key = lat * 1000 + lon
                if key in blocks:
                    blocks[key].append(i)
                else:   
                    blocks[lat * 1000 + lon] = [i]        
            except ValueError:
                print('value')
            except KeyError:
                print('v2')

        for key in blocks.keys():
            rows = blocks[key]
            lat = int(key / 1000)
            lon = key % 1000

            path = './output/'+str(zone)+"/"+str(year) +"/"+str(month)+"/"+str(day)+"/"+str(hour)+"/"+str(minute)+"/"+str(lat)+"/"+str(lon)
            if not os.path.exists(path):
                os.makedirs(path)

            with open(path + '/interactions.csv','w') as file:    
                for i in rows:   
                    outputRow = str(tb['BaseDateTime'][i]) + ',' + str(tb["MMSI"][i]) +','+str(tb["LAT"][i])+','+str(tb["LON"][i])+','+str(tb["SOG"][i])+','+str(tb["COG"][i])+'\n'
                    file.write(outputRow)       
```
